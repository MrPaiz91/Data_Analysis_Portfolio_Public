# Preparacion para Propension

```python
clientes_final['country_id'].value_counts()
```

```python
df_country_code2 = clientes_final['country_id'].value_counts(normalize=True).to_frame().reset_index()
df_country_code2.columns = ['country_id', 'frecuencia']
df_country_code2.head()
#Se hara otra columna con Lambda para dejar Espaa: 1 Otros: 0
```

```python
clientes_final['es_espaa?'] = clientes_final['country_id'].apply(lambda x:
                                                        1 if x == 'ES'
                                                        else 0)
```

```python
clientes_final[['es_espaa?']].value_counts()
```

```python
clientes_final.head(3)
```

```python
clientes_final.drop('country_id', axis=1, inplace=True)
clientes_final.head(3)
```

```python
clientes_final['es_espaa?'].value_counts()
```

```python
#Cambiando columnas con fecha de 'object' a 'datetime'
clientes_final['pk_partition'] = pd.to_datetime(clientes_final['pk_partition'])
clientes_final['entry_date'] = pd.to_datetime(clientes_final['entry_date'])
clientes_final['month_sale'] = pd.to_datetime(clientes_final['month_sale'])
```

```python
clientes_final.info()
```

```python
clientes_final['product_id'].value_counts()
```

```python
clientes_final['product_description'].value_counts()
# Ambas columnas tienen las mismas infos. Mejor elimino esta columna
```

```python
clientes_final.drop('product_description', axis=1, inplace=True)
```

```python
clientes_final.info()
```

```python
#One hot encodings
clientes_final = pd.get_dummies(clientes_final, columns=["age_group", "gender", "family_product"])
```

```python
clientes_final.info()
```

```python
clientes_final['segment'].value_counts()
#Voy a eliminar 'SIN SEGMENTO' para dejar solo 3 etiquetas a predecir
```

```python
clientes_final.drop(clientes_final[clientes_final.segment == '04 - SIN SEGMENTO'].index, inplace=True)
```

```python
clientes_final['segment'].value_counts()
```

```python
dict = {
    '01 - TOP': 1,
    '02 - PARTICULARES': 2,
    '03 - UNIVERSITARIO': 3
    }

clientes_final['segment'] = clientes_final['segment'].replace(dict)
clientes_final['segment'].value_counts()
```

```python
#One hot encoding para 'product_id'
clientes_final = pd.get_dummies(clientes_final, columns=['product_id'])
```

```python
clientes_final.info()
```

```python
clientes_final.drop(['pk_partition', 'entry_date', 'month_sale'], axis=1, inplace=True)
```

```python
clientes_final.info()
```
